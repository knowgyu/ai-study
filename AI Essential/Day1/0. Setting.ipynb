# 환경 변수 설정
### ChatGPT API 활용을 위한 환경 변수 설정
* `PROXY_URL`, `PROXY_TOKEN` 설정 및 변경
* ChatGPT 서버가 프록시로 동작하므로 해당 코드를 이용해서 `.env` 파일 내용을 업데이트합니다.

```python
# 파일 경로를 객체처럼 다루기 위한 Path를 불러옵니다.
from pathlib import Path

# .env 파일의 값을 수정하기 위한 set_key 함수를 불러옵니다.
from dotenv import set_key

# 현재 작업 폴더 기준으로 두 단계 상위 폴더의 .env 파일 경로를 설정합니다.
env_path = Path('.').resolve().parent.parent / ".env"

# .env 파일에 저장할 환경변수들을 딕셔너리로 정의합니다.
env_updates = {
    "PROXY_URL": "http://10.0.4.77:8000/v1",
    "PROXY_TOKEN": "ab0bb89cba5b86a4ce001032919fbefa"
}

# .env 파일이 없으면 새로 생성합니다.
if not env_path.exists():
    env_path.touch()

# 딕셔너리에 있는 각 환경변수를 하나씩 .env 파일에 기록합니다.
for key, value in env_updates.items():
    set_key(
        dotenv_path=env_path,
        key_to_set=key,
        value_to_set=value,
        export=True  # export KEY=value 형태로 저장합니다.
    )

# 설정이 완료되었음을 사용자에게 출력합니다.
print(f"✅ {env_path} 파일에 프록시 설정이 성공적으로 업데이트되었습니다.")
```
